**Comandos y Sintaxis Destacados (TUPLAS)**

* `,` (Coma) y `()` (Paréntesis): Permiten crear tuplas separando valores. Los paréntesis son opcionales, pero para crear una tupla de un solo elemento es obligatorio incluir una coma al final (ej. `t1 = 'p',`).

* `tuple()`: Función integrada para crear una tupla vacía o para convertir una secuencia (como un string o una lista) en una tupla con sus elementos.

* `divmod()`: Función integrada que toma dos números y devuelve una tupla de dos valores: el cociente y el resto.

* Operador de empaquetado/desempaquetado: Si se usa en la definición de parámetros de una función (`*args`), empaqueta múltiples argumentos variables en una sola tupla. Si se usa al invocar una función (`divmod(*t)`), desempaqueta una tupla para pasar sus elementos como argumentos separados.

* `zip()`: Función integrada que empareja elementos de dos o más secuencias y devuelve un objeto zip, útil para iterar secuencias simultáneamente.

* `enumerate()`: Función que recorre una secuencia devolviendo pares con el índice (empezando desde 0) y el elemento correspondiente.

* `items()`: Metodo de un diccionario que devuelve una vista de los pares clave-valor del diccionario como una lista de tuplas. Cada tupla contiene un par (clave, valor).

* `sorted(t,key,reverse=booleano)`, `max(t,key)`, `min(t,key)`: Estas funciones admiten un argumento opcional llamado `key` donde se puede pasar una función personalizada que determine cómo calcular el orden o la comparación.

* `structshape()`: Función de un módulo externo recomendado por el autor que toma una estructura de datos y devuelve un string resumiendo su tipo y composición para facilitar la depuración.


```python
mi_diccionario = {
    "nombre": "Ana",
    "edad": 30,
    "ciudad": "Madrid"
}

# Usamos el método items()
parejas = mi_diccionario.items()
print(parejas)

# Puedes iterar sobre estas parejas de clave-valor
for item in mi_diccionario.items():
  clave, valor = item
  print(f"La clave es '{clave} y el valor es '{valor}'")
```

**LECCIONES VALIOSAS Y CONCEPTOS CLAVE**

* Inmutabilidad y Hashabilidad: Las tuplas son secuencias como las listas, pero son inmutables, lo que significa que no soportan reasignación ni métodos como append o remove. Gracias a ser inmutables, las tuplas son "hashables" y pueden utilizarse como claves en los diccionarios, a menos que contengan en su interior valores mutables como listas.

* Asignación de Tuplas: Es posible poner una tupla de variables a la izquierda de una asignación y una secuencia de valores a la derecha. Esto resulta muy útil para intercambiar valores entre dos variables sin necesidad de utilizar una variable temporal (ej. a, b = b, a).

* Tuplas como Valores de Retorno: Aunque una función estrictamente devuelve un solo valor, al devolver una tupla el efecto práctico es el de retornar múltiples valores de manera simultánea.

* Comparación de Tuplas: Los operadores relacionales (<, >, etc.) comparan las tuplas evaluando el primer elemento de cada secuencia. Si son iguales, pasan al siguiente par y así sucesivamente hasta encontrar una diferencia. Esta lógica es la que usa Python por defecto al ordenar listas de tuplas.

* Inversión de Diccionarios: Al invertir un diccionario para buscar una clave a partir de un valor, surge el problema de que los valores originales pueden estar duplicados. La solución teórica es crear un diccionario nuevo donde los valores sean listas de claves que compartían el mismo valor original.

* Depuración de Estructuras Compuestas: A medida que los programas crean "estructuras de datos compuestas" (como diccionarios que contienen tuplas y listas), es fácil cometer errores de tipo, tamaño o formato. El autor sugiere evitar revisar datos voluminosos a mano y usar herramientas como structshape para inspeccionar visualmente de qué están compuestos los datos.

**Comparing and Sorting**

The relational operators work with tuples and other sequences.
For example, if you use the `<` operator with tuples, it starts by comparing the first element from each sequence.
If the relation is true, it goes on to the next pair of elements, and so on, until it finds a pair for which the relation is false.

```python
print((0.5, 2, 10) > (0, 20, 100)) #Decide en base a primer elemento
print((0, 2, 10) > (0, 20, 1)) #Como 1er elementos son iguales, entonces decide con 2do elementos
print((0, 20, 10) > (0, 2, 1)) #Decide con 2do elementos
```

This way of comparing tuples is useful for sorting a list of tuples, or finding the minimum or maximum.
As an example, let's find the most common letter in a word.
In the previous chapter, we wrote `value_counts`, which takes a string and returns a dictionary that maps from each letter to the number of times it appears.

```python
def value_counts(string):
    counter = {}
    for letter in string:
        if letter not in counter:
            counter[letter] = 1
        else:
            counter[letter] += 1
    return counter
```

```python
def count_sourting(string):
  t = value_counts(string)
  # Ordenar los ítems del diccionario por el segundo elemento de cada tupla (la frecuencia)
  return sorted(t.items(), key=lambda item: item[1])
```

```python
print(count_sourting("banana"))
```

In this chapter we made a dictionary that maps from each letter to its index in the alphabet. We can use letter_map and letters to encode and decode words using a Caesar cipher.

A Caesar cipher is a weak form of encryption that involves shifting each letter by a fixed number of places in the alphabet, wrapping around to the beginning if necessary. For example, 'a' shifted by 2 is 'c' and 'z' shifted by 1 is 'a'.

Write a function called shift_word that takes as parameters a string and an integer, and returns a new string that contains the letters from the string shifted by the given number of places.

```python
def invert_dict(d):
    new = {}
    for key, value in d.items():
        if value not in new:
            new[value] = [key]
        else:
            new[value].append(key)
    return new
```

```python
def shift_word(word, n):
    """Shift the letters of `word` by `n` places.

    >>> shift_word('cheer', 7)
    'jolly'
    >>> shift_word('melon', 16)
    'cubed'
    """
    alphabet = 'abcdefghijklmnopqrstuvwxyz'
    number_to_letter = dict(enumerate(alphabet))
    letter_to_number = invert_dict(number_to_letter)
    cipher_list = []
    word.lower()
    for letter in word:
      cipher_value = letter_to_number[letter]
      cipher_letter = number_to_letter[(cipher_value[0] + n) % len(alphabet)]
      cipher_list.append(cipher_letter)
    cipher_word="".join(cipher_list)
    return cipher_word

print(shift_word('cheer', 7))
print(shift_word('melon', 16))
```
