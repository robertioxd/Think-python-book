Write a function called `triangle` that takes a string and an integer and draws a pyramid with the given height, made up using copies of the string. Here's an example of a pyramid with `5` levels, using the string `'L'`.

```python
def triangle(str, n): #n es el numero de niveles
    for i in range(1, n + 1): #"n+1" porque range(1,n+1) itera en 1, 2, ..., n-1, n
        # Calculamos los espacios necesarios para centrar la pirámide
        espacios = ' ' * (n - i)
        # Imprimimos la línea: espacios + caracteres (usamos 2*i-1 para forma de punta)
        print(espacios + str * (2 * i - 1))
triangle("L",5)
```

A **docstring** is a string at the beginning of a function that explains the interface ("doc" is short for "documentation").
Here is an example:

```python
def polyline(n, length, angle):
    """Draws line segments with the given length and angle between them.

    n: integer number of line segments
    length: length of the line segments
    angle: angle between segments (in degrees)
    """
    for i in range(n):
        forward(length)
        left(angle)
polyline
```

By convention, docstrings are triple-quoted strings, also known as **multiline strings** because the triple quotes allow the string to span more than one line.

A docstring should:

* Explain concisely what the function does, without getting into the details of how it works,

* Explain what effect each parameter has on the behavior of the function, and

* Indicate what type each parameter should be, if it is not obvious.

Writing this kind of documentation is an important part of interface design.
A well-designed interface should be simple to explain; if you have a hard time explaining one of your functions, maybe the interface could be improved.

El propósito principal de return es enviar un dato desde el interior de la función hacia el lugar del código donde esa función fue llamada (invocada). Ese valor devuelto puede ser guardado en una variable o usado en otra operación.

```python
def calcular_cuadrado(numero):
    resultado = numero * numero
    return resultado  # Aquí entregamos el valor

# Llamamos a la función y GUARDAMOS lo que nos "returna" en una variable
mi_numero = calcular_cuadrado(4)
print(mi_numero)  # Esto imprimirá 16
```

En el momento en que Python lee la palabra return dentro de una función, la función se detiene inmediatamente. Cualquier código que esté escrito debajo de ese return (en el mismo bloque) jamás se ejecutará.

`isinstance(element, type)` es una funcion que admite 2 argumentos y permite verificar que un elemento sea de un tipo, devolviendo `True` si el elemento correspodne al tipo impuesto  y dando `False` si corresponde.

```python
def factorial(n):
    if not isinstance(n, int):
        print('factorial is only defined for integers.')
        return None
    elif n < 0:
        print('factorial is not defined for negative numbers.')
        return None
    elif n == 0:
        return 1
    else:
        return n * factorial(n-1)
factorial('crunchy frog')
```

Métodos del módulo `re`

* `re.search()`: Busca un patrón de expresión regular a lo largo de toda una cadena. Si encuentra una coincidencia, devuelve un objeto especial (Match object); si no, devuelve None. Es una versión mucho más poderosa que el método find(). Por ejemplo, `re.search('^A', 'Apple')` busca si la cadena empieza exactamente con 'A'.
* `re.sub()`: Encuentra todas las coincidencias de un patrón de expresión regular en una cadena y las sustituye por un nuevo texto. Es la versión avanzada e inteligente del método `replace()`. Por ejemplo, `re.sub('\s+', ' ', 'texto   con   espacios')` reemplaza múltiples espacios por un solo espacio.

```python
def has_e(word):
    for letter in word.lower():
        if letter == 'e':
            return True
    return False

def has_e(word): # Version de "has_e" usando "in"
    return 'e' in word.lower()

def uses_any(word,letters): # Version mas general de "has_e". "letters" son el remplazo de "e"
  for letter in letters:
    if letter in word.lower():
      return True
  return False

def uses_all(s1,s2): #Checkea si todo caracter de s2 se encuentra en s1
  for char in s2.lower():
    if not uses_any(s1,char):
      return False
  return True
```

El aliasing (o creación de alias) ocurre cuando más de una variable hace referencia al mismo objeto. Dado que un objeto con múltiples referencias tiene más de un nombre, se dice que el objeto está "aliased" (tiene un alias). Esto sucede típicamente de dos maneras:

* Por asignación directa: Si una variable a hace referencia a un objeto y ejecutas b = a, ambas variables ahora apuntan exactamente al mismo objeto en la memoria.
* A través de funciones: Cuando pasas la referencia de un objeto como argumento a una función, se crea una forma de aliasing entre la variable original y el parámetro de la función.

El impacto del aliasing depende completamente de si la variable afectada es mutable o inmutable:

* En objetos mutables (como las listas): Si el objeto con alias es mutable, los cambios realizados con un nombre afectarán al otro. Por ejemplo, si una función modifica una lista que recibió como argumento, el código original que llamó a la función también verá reflejado ese cambio. Aunque esto puede resultar útil, es muy propenso a generar errores, por lo que se recomienda evitar el aliasing al trabajar con objetos mutables.
* En objetos inmutables (como los strings): El aliasing no representa un problema. Como su contenido no puede ser modificado tras su creación, casi nunca importa si dos o más variables hacen referencia a la misma cadena de texto.

**List arguments**

When you pass a list to a function, the function gets a reference to the
list. If the function modifies the list, the caller sees the change. For
example, `pop_first` uses the list method `pop` to remove the first element from a list.

```python
def pop_first(lst):
    return lst.pop(0)
letters = ['a', 'b', 'c']
pop_first(letters) #modifica la lista letters
print(letters)
```
