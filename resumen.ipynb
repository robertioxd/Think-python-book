Write a function named `print_right` that takes a string named `text` as a parameter and prints the string with enough leading spaces that the last letter of the string is in the 40th column of the display.

```python
def print_right(text):
    n=40-len(text)
    print(' '*n+text)
print_right("Monty")
print_right("Python's")
print_right("Flying Circus")
```

Write a function called `triangle` that takes a string and an integer and draws a pyramid with the given height, made up using copies of the string. Here's an example of a pyramid with `5` levels, using the string `'L'`.

```python
def triangle(str, n): #n es el numero de niveles
    for i in range(1, n + 1): #"n+1" porque range(1,n+1) itera en 1, 2, ..., n-1, n
        # Calculamos los espacios necesarios para centrar la pirámide
        espacios = ' ' * (n - i)
        # Imprimimos la línea: espacios + caracteres (usamos 2*i-1 para forma de punta)
        print(espacios + str * (2 * i - 1))
triangle("L",5)
```

A **docstring** is a string at the beginning of a function that explains the interface ("doc" is short for "documentation").
Here is an example:

```python
def polyline(n, length, angle):
    """Draws line segments with the given length and angle between them.

    n: integer number of line segments
    length: length of the line segments
    angle: angle between segments (in degrees)
    """
    for i in range(n):
        forward(length)
        left(angle)
polyline
```

By convention, docstrings are triple-quoted strings, also known as **multiline strings** because the triple quotes allow the string to span more than one line.

A docstring should:

* Explain concisely what the function does, without getting into the details of how it works,

* Explain what effect each parameter has on the behavior of the function, and

* Indicate what type each parameter should be, if it is not obvious.

Writing this kind of documentation is an important part of interface design.
A well-designed interface should be simple to explain; if you have a hard time explaining one of your functions, maybe the interface could be improved.

El propósito principal de return es enviar un dato desde el interior de la función hacia el lugar del código donde esa función fue llamada (invocada). Ese valor devuelto puede ser guardado en una variable o usado en otra operación.

```python
def calcular_cuadrado(numero):
    resultado = numero * numero
    return resultado  # Aquí entregamos el valor

# Llamamos a la función y GUARDAMOS lo que nos "returna" en una variable
mi_numero = calcular_cuadrado(4)
print(mi_numero)  # Esto imprimirá 16
```

En el momento en que Python lee la palabra return dentro de una función, la función se detiene inmediatamente. Cualquier código que esté escrito debajo de ese return (en el mismo bloque) jamás se ejecutará.

```python
def evaluar_clima(temperatura):
    if temperatura > 30:
        return "Hace mucho calor" # Si entra aquí, la función termina.
    return "El clima es agradable"
    print("Esto nunca se va a imprimir en la pantalla") # Código inalcanzable
evaluar_clima(30.0000001)
```

Una de las cosas geniales de Python es que return te permite devolver varios datos al mismo tiempo, separados por comas. Internamente, Python los agrupa en una estructura llamada "tupla".

```python
def obtener_datos_usuario():
    nombre = "Ana"
    edad = 28
    ciudad = "Madrid"
    return (nombre, edad, ciudad) # Devolvemos tres cosas de golpe, puede ser sin los parentesis

# Podemos atrapar esos tres valores en tres variables distintas:
a = (n, e, c) = obtener_datos_usuario()
print(a)
```

`isinstance(element, type)` es una funcion que admite 2 argumentos y permite verificar que un elemento sea de un tipo, devolviendo `True` si el elemento correspodne al tipo impuesto  y dando `False` si corresponde.

```python
def factorial(n):
    if not isinstance(n, int):
        print('factorial is only defined for integers.')
        return None
    elif n < 0:
        print('factorial is not defined for negative numbers.')
        return None
    elif n == 0:
        return 1
    else:
        return n * factorial(n-1)
factorial('crunchy frog')
```
